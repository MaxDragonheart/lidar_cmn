# Download dei dati

## Dati iniziali

```python
from pathlib import Path

geodata_path = Path.cwd().parent.joinpath("geodata")
limiti_amministrativi = geodata_path.joinpath("limiti_amministrativi.gpkg")
quadro_unione = geodata_path.joinpath("quadro_unione_lidar.gpkg")
lidar_type = ["dtm", "dsm"]

target_area_name = "Procida"

output_folder = Path("/home/max/Desktop/lidar")
output_folder.exists()
```

## Selezione e Download

```python
from lidar_cmn.vector import read_vector, target_area


for lidar in lidar_type:
    base = read_vector(data_path=quadro_unione, layer_name=lidar)
    ax = base.plot(alpha=0.75, color="blue", figsize=(10, 10))
    ax.set_title(lidar)

    data = target_area(
        boundaries=read_vector(data_path=limiti_amministrativi),
        link_source=read_vector(data_path=quadro_unione, layer_name=lidar),
        target=target_area_name
    )
    data.plot(ax=ax, alpha=0.75, color="yellow")
```

### DTM

```python

from lidar_cmn.download import download_data

dsm = target_area(
    boundaries=read_vector(data_path=limiti_amministrativi),
    link_source=read_vector(data_path=quadro_unione, layer_name=lidar_type[0]),
    target=target_area_name
)

download_data(
    selection=dsm,
    output_folder=output_folder,
    output_folder_name=lidar_type[0]
)
```

### DSM

```python
dsm = target_area(
    boundaries=read_vector(data_path=limiti_amministrativi),
    link_source=read_vector(data_path=quadro_unione, layer_name=lidar_type[1]),
    target=target_area_name
)

download_data(
    selection=dsm,
    output_folder=output_folder,
    output_folder_name=lidar_type[1]
)
```
